```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

```python
df = pd.read_csv('/content/drive/MyDrive/zomato.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df = df.drop(['url', 'address', 'phone', 'menu_item', 'dish_liked', 'reviews_list'], axis = 1)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.shape
```

```python
df['rate'].unique()
```

```python
def handlerate(value):
  if(value == 'NEW' or value == '-'):
    return np.nan
  else:
    value = str(value).split('/')
    value = value[0]
    return float(value)

df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

```python
df['rate'].isnull().sum()
```

```python
df['rate'].fillna(df['rate'].mean(), inplace = True)
df['rate'].isnull().sum()
```

```python
df.dropna(inplace = True)
df.info()
```

```python
df.rename(columns = {'approx_cost(for two people)': 'cost2people', 'listed_in(type)': 'type'}, inplace = True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

```python
df = df.drop(['listed_in(city)'], axis = 1)
```

```python
df['cost2people'].unique()
```

```python
def handlecomma(value):
  value = str(value)
  if ',' in value:
    value = value.replace(',', '')
    return float(value)
  else:
    return float(value)

df['cost2people'] = df['cost2people'].apply(handlecomma)
df['cost2people'].head()
```

```python
df['rest_type'].value_counts()
```

```python
df['rest_type'].unique()
```

```python
rest_types = df['rest_type'].value_counts(ascending = True)
rest_types
```

```python
resttype_lessthan1000 = rest_types[rest_types < 1000]
resttype_lessthan1000
```

```python
def handle_rest_type(value):
  if(value in resttype_lessthan1000):
    return 'others'
  else:
    return value

df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df['location'].value_counts()
```

```python
location_arrange = df['location'].value_counts(ascending = True)
location_arrange
```

```python
location_lessthan300 = location_arrange[location_arrange < 300]
location_lessthan300
```

```python
def handle_location(value):
  if value in location_lessthan300:
    return 'others'
  else:
    return value

df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
df.columns
```

```python
df['type'].unique()
```

```python
cuisines_arrange=df['cuisines'].value_counts(ascending=True)
cuisines_arrange
```

```python
cuisines_lessthan100=cuisines_arrange[cuisines_arrange<100]
cuisines_lessthan100
```

```python
def handle_cuisines(value):
  if value in cuisines_lessthan100:
    return 'others'
  else:
    return value

df['cuisines']=df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
